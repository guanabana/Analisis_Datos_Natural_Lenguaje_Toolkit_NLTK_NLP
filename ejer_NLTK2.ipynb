```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import requests
```

```python
# URL de la pÃ¡gina web que deseas analizar
#url = 'https://web.comisiondelaverdad.co/actualidad/noticias/principales-cifras-comision-de-la-verdad-informe-final'
url = 'https://web.comisiondelaverdad.co/actualidad/noticias/ejecuciones-extrajudiciales-trabajo-esclarecimiento-comision-verdad'

# Realizar la solicitud HTTP y obtener el contenido HTML de la pÃ¡gina
response = requests.get(url)
html = response.text

# Parsear el HTML utilizando BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Extraer el texto del cuerpo de la pÃ¡gina
texto_pagina = ""
for parrafo in soup.find_all('p'):
    texto_pagina += parrafo.get_text() + " "

# TokenizaciÃ³n y eliminaciÃ³n de stop words
tokens = word_tokenize(texto_pagina)
stop_words = set(stopwords.words('spanish'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Unir las palabras en un solo string
texto_procesado = " ".join(filtered_tokens)

# Crear la nube de palabras
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stop_words, 
                min_font_size = 10).generate(texto_procesado)

# Visualizar la nube de palabras
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```
